Q2:CAL. the number of tweets and number of retweets
To get the this detail we look at the "text" column anyone that starts with "RT" is the retweet ones otherwise it is the main tweet

```python
tweets = [line for line in df.text if not line.startswith("RT")]
retweets = [ line for line in df.text if line.startswith("RT")]

len(retweets) + len(tweets) == len(df)
```

```python
len(df)
```

```python
len_tweets = len(tweets)
len_retweets = len(retweets)

labels = "tweets","retweets"
sizes = [len_tweets, len_retweets]
colors = ["green","red"]

explode = (0.1, 0)
plt.pie(sizes,explode=explode,labels = sizes, autopct = "%1.1f%%",colors =colors)
plt.title("Number of tweets and number of retweets")
plt.legend()
plt.show()
```

Q3:Cal. which 5 people did they retweet their posts the most and how many each

```python
tweets[:5]
```

```python
retweets[:5]
```

```python
#extract the first mention in each retweet
retweetees = []
for line in retweets:
    for item in line:
        if item.startswith("@"):
            x = line.split()[1]
            x = x.replace(":","")
            retweetees.append(x)
            break

retweetees2 = [line.split()[1].replace(":","")for line in retweets]
```

```python
freq = Counter(retweetees)
```

```python
sorted_outcome = [(mention,amount) for mention,amount in sorted([(x,y) for y,x in freq.items()],reverse = True)]
sorted_outcome[:5]
```

```python
print(len(retweetees) == len(retweets))
print(len(retweetees2) == len(retweets))
print(len(retweetees2[:5])== len(retweetees[:5]))
```

```python
retweetees[:5]
```

```python
retweetees_df = pd.DataFrame()
retweetees_df["mentions"] = retweetees

retweetees_df.mentions.value_counts().head()
```

```python
mentions = [line[0] for line in sorted_outcome[:5]][::-1]
y_pos = range(len(mentions))
frequency = [line[1] for line in sorted_outcome[:5]][::-1]

plt.barh(y_pos, frequency)
plt.yticks(y_pos, mentions)

plt.xlabel("frequency")
plt.ylabel("Retweetees")
plt.title("Top 5 users retweet by X")

plt.show()
```
